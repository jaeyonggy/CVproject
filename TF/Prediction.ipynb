### Prediction

```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing import image
from tensorflow.keras.utils import image_dataset_from_directory
import numpy as np
import os
from tqdm import tqdm
```

Loading the latest save model after fine tuning

```python
model_folder = 'effV2B0'
csv_name = 'effV2B0_TF.csv'
IMG_SIZE = 300
```

```python
best_model = keras.models.load_model('/content/drive/MyDrive/ZINDI-image/checkpoint/'+model_folder)
```

Loading test images

```python
folder_path = '/content/drive/MyDrive/ZINDI-image/test'
pred = []

# for each test image
for img in tqdm(sorted(os.listdir(folder_path))):
    img = os.path.join(folder_path, img)
    img = image.load_img(img, target_size=(IMG_SIZE, IMG_SIZE))
    img = image.img_to_array(img)  # image array
    img = np.expand_dims(img, axis=0)  # image batch
    # predicting the labels
    pred.extend(best_model.predict(img))
```

```python
pred_labels = []
for i in pred:
  if i[0] <= 0.5:
    pred_labels.append(0)
  else:
    pred_labels.append(1)
```

```python
submission = pd.read_csv('/content/drive/MyDrive/ZINDI-image/csv_files/SampleSubmission.csv')

# Replacing the labels with the predicted ones
submission['Label'] = pred_labels

submission
```

```python
# Saving it as a csv
submission.to_csv('/content/drive/MyDrive/ZINDI-image/csv_files/'+csv_name, index=False)
```
