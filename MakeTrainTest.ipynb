```python
import os
import shutil
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/ZINDI-image/csv_files/Train.csv')  # train set

# traverse whole directory
for root, dirs, files in os.walk('/content/drive/MyDrive/ZINDI-image/Images'):
    # select file name
    for file in files:
        if file in list(df['Image_id']):  # train
          shutil.move(root+'/'+file, '/content/drive/MyDrive/ZINDI-image/train/'+file)
        else:  # test
          shutil.move(root+'/'+file, '/content/drive/MyDrive/ZINDI-image/test/'+file)
```
