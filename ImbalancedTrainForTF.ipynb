```python
import os
import shutil
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/ZINDI-image/csv_files/New_Train.csv')  # train set
df
```

```python
# traverse whole directory
for root, dirs, files in os.walk('/content/drive/MyDrive/ZINDI-image/imbalanced_train/'):
    # select file name
    for file in files:
        if df[df['Image_id']==file].iloc[0]['Label'] == 0:
          shutil.copy(root+file, '/content/drive/MyDrive/ZINDI-image/imbalanced_train_TF/0/'+file)
        else:
          shutil.copy(root+file, '/content/drive/MyDrive/ZINDI-image/imbalanced_train_TF/1/'+file)
```

```python
count = 0

for root, dirs, files in os.walk('/content/drive/MyDrive/ZINDI-image/imbalanced_train_TF/0'):
    # select file name
    for file in files:
        if file.endswith('.jpg'):
            count += 1

count
```
