```python
import os
import shutil
import pandas as pd
import random
```

```python
df = pd.read_csv('/content/drive/MyDrive/ZINDI-image/csv_files/Train.csv')
df  # 1: Damaged
```

```python
normal_idx = df.index[df['Label'] == 0].tolist()
len(normal_idx)
```

```python
damaged_idx = df.index[df['Label'] == 1].tolist()
len(damaged_idx)
```

```python
# Making imbalanced data by only selecting the amount of damaged leaves by 10% of the number of normal leaves
random.seed(0)  # setting the seed
imbalanced_idx = sorted(random.sample(damaged_idx, int(len(normal_idx)/10)))
count = 0
for item in imbalanced_idx:
  count += 1
  print(item, end=" ")
  if count % 10 == 0:
    print()

print("\n\nTotal:", count)
```

```python
# Making a new train csv file
new_train_df = df[df['Label']==0]  # first getting the normal leaves only
new_train_df = pd.concat([new_train_df,df.iloc[imbalanced_idx]])  # adding the selected damaged leaves
new_train_df = new_train_df.sort_values('Image_id')  # sorting by the Image_id column
new_train_df = new_train_df.reset_index(drop=True)  # resetting the index
new_train_df.to_csv('/content/drive/MyDrive/ZINDI-image/csv_files/New_Train.csv')
new_train_df
```

```python
# Making the imbalanced train dataset
df = pd.read_csv('/content/drive/MyDrive/ZINDI-image/csv_files/New_Train.csv')  # train set

# traverse whole directory
for root, dirs, files in os.walk('/content/drive/MyDrive/ZINDI-image/train'):
    # select file name
    for file in files:
        if file in list(df['Image_id']):  # train
          shutil.copy(root+'/'+file, '/content/drive/MyDrive/ZINDI-image/imbalanced_train/'+file)
```
